```python
data = []

for idx, row in df.iterrows():
    if idx < 302:
        continue
    print(idx, '/', len(df))
    link = row['Course Link']
    driver.get(link)
    time.sleep(1)
    
    tabs = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul[@id='myTab']/li/a"))
    )
    # Initialize an empty list to store the data
    # Loop through each tab and click on it
    for tab in tabs:
        # Get the tab text
        tab_text = tab.text
        # Click on the tab
        tab.click()
        time.sleep(1)   
    
    # Get the page source from the Selenium driver
    page_source = driver.page_source
    # Parse the page source
    soup = BeautifulSoup(page_source, 'html.parser')

    data_dict = {'Course Link': link}
    # Extract the key-value pairs
    all_dl = soup.find_all('dl', class_='c-description-list c-description-list--striped')
    for dl in all_dl:
        for dt, dd in zip(dl.find_all('dt'), dl.find_all('dd')):
            key = dt.get_text(strip=True)
            value = dd.get_text(strip=True)
            data_dict[key] = value
        
    # Extract website
    try:
        website_element = soup.find('a', class_='visitCourseWebsite')
        if website_element:
            website_url = website_element['href']
            data_dict['Website'] = website_url

        # Extract email contact
        email_element = soup.find('a', class_='writeEmail')
        if email_element:
            email_address = email_element['href'].replace('mailto:', '')
            data_dict['Email'] = email_address
    except:
        data_dict['Website'] = None
        data_dict['Email'] = None
        
    
    data.append(data_dict)

pd.DataFrame(data).to_csv('/home/sanket/Documents/daad_germany_scraped_internal2.csv', index=None)
```
